# Python Randomly Generated Workout Plan

This code is used for the purpose of generating a **randomized workout plan**.
Sometimes I want to mix things up at the gym, so I created this randomizer code that allows me to have randomized workouts generated by python without any decisions to be done on my end. 


The creation of the code, allowed me to brush up on my python skills expecially on the creation of **loops**. 

# Understang the code 

The workouts are generated through the use of a rolled dice that has a workout attributed to each number. <br>
The dice will be **rolled the amount of different workouts you want** to create a whole workout plan.
- 1 : Leg work out
- 2 : Chest work out
- 3 : Back work out
- 4 : Cardio 
- 5 : Abs workout 
- 6 : Arm work out 

```python
# Import the randomizer package

import numpy as np

# Defining the function dice and its variables  
def dice():
    nb_rolls = input( "Enter the amount of workouts you want: " )
    nb_rolls = int( nb_rolls )
    injury = input( "Enter any injury: " )
    workout_plan = []
    noplan= True

# We add 'noplan = True' so that the dice rerolls if it falls on a number assigned to an injury 
# If we didn't add 'noplan' than when the dice falls on a number corresponding to an injury, the dice would continue rolling
# This is not what we want
# To better explain the problem: if I tell the algorithm that my leg is injured (dice = 1) and I ask the dice to roll out 3 times 
# and the dice falls on 1 but the 2 other times the dice doesnt, I will only end up with a workout plan with 2 work outs
# but I want a workout plan with 3 workouts, that is why I asked for it to be rolled 3 times 
# So we must add the while loop to check for 'noplan = false'
# I wrote 'noplan' just because the workout plan hasn't been completed yet 
    

    # Creating the 'for loop' with 'if' conditions so the algorithm checks the dice number then prints out the type of workout
    # X is in the range 'nb_rolls' to make sure the dice is rolled as many times as the user asked for workouts
    # != is used to make sure the algorithm does not output a workout that uses an injured body part 
    for x in range(nb_rolls):
        noplan = True
        while noplan:
            dice = np.random.randint(1, 7)
            if dice == 1 and injury != 'leg':
                workout_plan.append("Leg ")
                noplan = False
            elif dice == 2 and injury != 'chest':
                workout_plan.append("Chest  ")
                noplan = False
            elif dice == 3 and injury != 'back':
                workout_plan.append("Back ")
                noplan = False
            elif dice == 4 and injury != 'leg':
                workout_plan.append("Cardio ")
                noplan = False
            elif dice == 5 and injury != 'abs':
                workout_plan.append("Abs ")
                noplan = False
            elif dice == 6 and injury != 'arm':
                workout_plan.append("Arm ")
                noplan = False
    
    print("So your workout plan is :", ", " . join(workout_plan))
    print("Have a great workout !!")
            
dice()
```
